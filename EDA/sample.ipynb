```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys, os
this_path = '/home/ibi/Documents/GitHub/mas291-project/'
sys.path.append(this_path)
os.chdir(this_path)
print(os.getcwd())
```

```python
chicago = pd.read_csv('data/cleaned/chicago_sold_cleaned.csv')
chicago.head()
```

```python
count = chicago['community_area'].value_counts().sort_values(ascending=False)
count
```

```python
nyc = pd.read_csv('data/cleaned/new_york_sold_cleaned.csv')
count2 = nyc['community_area'].value_counts().sort_values(ascending=False)
count2
```

```python
top5_chicago = count.head(5).index
top5_nyc = count2.head(5).index

chicago_top5 = chicago[chicago['community_area'].isin(top5_chicago)]
nyc_top5 = nyc[nyc['community_area'].isin(top5_nyc)]
```

```python
chicago_top5.head()
```

```python
nyc_top5.head()
```

```python
selected_chicago = chicago_top5.select_dtypes(include=[np.number])
correlation_chicago = selected_chicago.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_chicago, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Top 10 Community Areas in Chicago')
plt.show()
```

```python
selected_nyc = nyc_top5.select_dtypes(include=[np.number])
correlation_nyc = selected_nyc.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_nyc, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Top 10 Community Areas in New York')
plt.show()
```

```python
chicago_selling = pd.read_csv('data/cleaned/chicago_selling_cleaned.csv')
nyc_selling = pd.read_csv('data/cleaned/new_york_selling_cleaned.csv')
count3 = chicago_selling['community_area'].value_counts().sort_values(ascending=False)
count4 = nyc_selling['community_area'].value_counts().sort_values(ascending=False)
count3, count4
```

```python
top5_chicago_selling = count3.head(5).index
top5_nyc_selling = count4.head(5).index

chicago_top5_selling = chicago_selling[chicago_selling['community_area'].isin(top5_chicago_selling)]
nyc_top5_selling = nyc_selling[nyc_selling['community_area'].isin(top5_nyc_selling)]

selected_chicago_selling = chicago_top5_selling.select_dtypes(include=[np.number])
correlation_chicago_selling = selected_chicago_selling.corr()

selected_nyc_selling = nyc_top5_selling.select_dtypes(include=[np.number])
correlation_nyc_selling = selected_nyc_selling.corr()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_chicago_selling, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Top 5 Community Areas in Chicago (Selling)')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_nyc_selling, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Top 5 Community Areas in New York (Selling)')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.scatterplot(x='area', y='price', color='blue', data=chicago_top5_selling)
plt.title('Scatter Plot for Top 5 Selling Community Areas in Chicago')
plt.show()
```

```python
nyc_top1 = count2.head(1).index
nyc_top1 = nyc[nyc['community_area'].isin(nyc_top1)]

selected_nyc_top1 = nyc_top1.select_dtypes(include=[np.number])
correlation_nyc_top1 = selected_nyc_top1.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_nyc_top1, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Top 1 Community Area in New York')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.scatterplot(x='area', y='price', color='red', data=nyc_top1, marker='x')
plt.title('Scatter Plot for Top 1 Community Area in New York')
plt.show()
```

```python
nyc_selling_top1 = count4.head(1).index
nyc_selling_top1 = nyc_selling[nyc_selling['community_area'].isin(nyc_selling_top1)]

selected_nyc_selling_top1 = nyc_selling_top1.select_dtypes(include=[np.number])
correlation_nyc_selling_top1 = selected_nyc_selling_top1.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_nyc_selling_top1, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Top 1 Community Area in New York (Selling)')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.scatterplot(x='area', y='price', color='red', data=nyc_selling_top1)
plt.title('Scatter Plot for Top 1 Community Area in New York (Selling)')
plt.show()
```
