```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
import sys, os
this_path = '/home/ibi/Documents/GitHub/mas291-project/'
sys.path.append(this_path)
os.chdir(this_path)
print(os.getcwd())
```

```python
chicago_selling = pd.read_csv("data/cleaned/chicago_selling_cleaned.csv")
chicago_sold = pd.read_csv("data/cleaned/chicago_sold_cleaned.csv")
new_york_selling = pd.read_csv("data/cleaned/new_york_selling_cleaned.csv")
new_york_sold = pd.read_csv("data/cleaned/new_york_sold_cleaned.csv")
```

```python
lat_chicago_selling = chicago_selling['latitude']
lon_chicago_selling = chicago_selling['longitude']
lat_chicago_sold = chicago_sold['latitude']
lon_chicago_sold = chicago_sold['longitude']
chicago_geojson = 'data/geojson/chicago_boundary.geojson'
chicago_comm_geojson = 'data/geojson/Chicago_community_area.geojson'

chicago_map = folium.Map(
    location=[lat_chicago_selling.mean(), lon_chicago_selling.mean()], 
    tiles='CartoDB positron'
)

marker_cluster = MarkerCluster(name='Houses on sale').add_to(chicago_map)

folium.GeoJson(
    chicago_geojson, 
    name='Boundary',
    style_function=lambda x: {
        'weight': 1
    }
).add_to(chicago_map)

folium.GeoJson(
    chicago_comm_geojson, 
    name='Community Area',
    style_function=lambda x: {
        'weight': 1
    },
    show=False
).add_to(chicago_map)

for i in range(len(chicago_selling)):
    folium.Marker(
        [lat_chicago_selling[i], lon_chicago_selling[i]], 
        popup=chicago_selling['address'][i],
        icon=folium.Icon(
            icon='home',
            color='blue'
        )
    ).add_to(marker_cluster)

marker_cluster2 = MarkerCluster(
    name='Houses sold', 
    show=False
).add_to(chicago_map)

for i in range(len(chicago_sold)):
    folium.Marker(
        [lat_chicago_sold[i], lon_chicago_sold[i]], 
        popup=chicago_sold['address'][i], 
        icon=folium.Icon(
            icon = 'home',
            color='green'
        )
    ).add_to(marker_cluster2)

folium.LayerControl().add_to(chicago_map)

sw_chicago_selling = chicago_selling[['latitude', 'longitude']].min().values.tolist()
ne_chicago_selling = chicago_selling[['latitude', 'longitude']].max().values.tolist()
chicago_map.fit_bounds([sw_chicago_selling, ne_chicago_selling])

chicago_map.save('helpers/map/chicago_map.html')

chicago_map
```

```python
chicago_choropleth = folium.Map(
    location=[lat_chicago_selling.mean(), lon_chicago_selling.mean()], 
    tiles='CartoDB positron'
)

folium.GeoJson(
    chicago_geojson, 
    name='Boundary',
    style_function=lambda x: {
        'weight': 1
    },
    show=False
).add_to(chicago_choropleth)

folium.GeoJson(
    chicago_comm_geojson, 
    name='Community Area',
    style_function=lambda x: {
        'weight': 1
    },
    show=False
).add_to(chicago_choropleth)

folium.Choropleth(
    geo_data=chicago_comm_geojson,
    name='Price',
    data=chicago_selling,
    columns=['community_area', 'price'],
    key_on='feature.properties.community',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Price'
).add_to(chicago_choropleth)

folium.Choropleth(
    geo_data=chicago_comm_geojson,
    name='Sold Price',
    data=chicago_sold,
    columns=['community_area', 'price'],
    key_on='feature.properties.community',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Sold Price',
    show=False
).add_to(chicago_choropleth)

folium.Choropleth(
    geo_data=chicago_comm_geojson,
    name='Area of selling houses',
    data=chicago_selling,
    columns=['community_area', 'area'],
    key_on='feature.properties.community',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Area of selling houses',
    show=False
).add_to(chicago_choropleth)

folium.Choropleth(
    geo_data=chicago_comm_geojson,
    name='Area of sold houses',
    data=chicago_sold,
    columns=['community_area', 'area'],
    key_on='feature.properties.community',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Area of sold houses',
    show=False
).add_to(chicago_choropleth)

folium.LayerControl().add_to(chicago_choropleth)

chicago_choropleth.fit_bounds([sw_chicago_selling, ne_chicago_selling])

chicago_choropleth.save('helpers/map/chicago_choropleth.html')

chicago_choropleth
```
